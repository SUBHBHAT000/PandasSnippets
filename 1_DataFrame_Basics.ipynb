## What is data frame ?
DataFrame is the heart of pandas library. It can be conceived as a table in any RDBMS or a spreadsheet having rows and columns used to represent tabular data.

# The equivalent SQL is

*SELECT * FROM temperature WHERE temperaturemin < 8;*

```python
# Or Equivalently
# Get the row which has the maximum temperaturemin
temp_df[temp_df['temperaturemin'] == temp_df['temperaturemin'].max()]
```

# The equivalent SQL is

*SELECT * FROM temperature 
 WHERE temperaturemin = (SELECT MAX(temperaturemin) FROM temperature);*

![SELECT * FROM temperature 
 WHERE temperaturemin = (SELECT MAX(temperaturemin) FROM temperature);](images/SQL_WHERE1.jpg "SELECT * FROM temperature 
 WHERE temperaturemin = (SELECT MAX(temperaturemin) FROM temperature);")

```python
# Get the row which has the maximum temperaturemin and select only specific columns
temp_df[['date_recorded','temperaturemin','temperaturemax']][temp_df.temperaturemin == temp_df.temperaturemin.max()]
```
